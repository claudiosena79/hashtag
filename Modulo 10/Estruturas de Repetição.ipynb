#### 1. Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

#### 3. Faça um programa que leia e valide as seguintes informações (e para cada uma delas, continue pedindo a informação até o usuário inserir corretamente):
##### Nome: maior que 3 caracteres;
##### Idade: entre 0 e 150;
##### Salário: maior que zero;
##### Sexo: 'f' ou 'm';
##### Estado Civil: 's', 'c', 'v', 'd';

````python
while True:
    nome = input('Digite nome:')
    if len(nome) < 4:
        continue
    else:
        break
while True:
    idade = int(input('Digite idade:'))
    if idade < 0 or idade > 150:
        continue
    else:
        break
while True:
    salario = float(input('Digite salario:'))
    if salario <= 0:
        continue
    else:
        break
while True:
    sexo = input('Digite sexo:')
    if sexo != 'f' and sexo != 'm':
        continue
    else:
        break
while True:
    estado_civil = input('Digite estado civil:')
    if estado_civil != 's' and estado_civil != 'c' and estado_civil != 'v' and estado_civil != 'd':
        continue
    else:
        break

print(f'NOME: {nome}, IDADE: {idade}, SALARIO: R${salario}, SEXO: {sexo}, ESTADO CIVIL: {estado_civil}')
````

#### 4. Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

````python
populacao_a = 80000
crescimento_a = 0.03
populacao_b = 200000
crescimento_b = 0.015
tempo = 0

while populacao_a < populacao_b:
    populacao_a = populacao_a * (1 + crescimento_a)
    populacao_b = populacao_b * (1+ crescimento_b)
    tempo += 1

print(f'A população de A ultrapassa a população de B em {tempo} anos')
````

#### 6. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o maior faturamento

````python
faturamentos =[]
i = 1
while True:
    if len(faturamentos) < 6:
        faturamento = float(input(f'Digite o faturamento do mês {i}'))
        faturamentos.append(faturamento)
        i += 1
    else:
        print(faturamentos)
        print(f'O maior faturamento foi R${max(faturamentos)}')
        break
````

#### 7. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento médio por mês (média).

````python
faturamentos =[]
i = 1
while True:
    if len(faturamentos) < 6:
        faturamento = float(input(f'Digite o faturamento do mês {i}'))
        faturamentos.append(faturamento)
        i += 1
    else:
        print(faturamentos)
        print(f'A soma dos faturamentos é de R${sum(faturamentos)}')
        print(f'A média dos faturamentos é de R${sum(faturamentos)/len(faturamentos)}')
        break
````

#### 9. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

````python
total_eleitores = int(input('Qual o total de eleitores: '))
votos = []
votacao_a = 0
votacao_b = 0
votacao_c = 0
for i in range(total_eleitores):
    voto = input('Digite seu voto: [A, B ou C]')
    votos.append(voto)
    if voto.upper() == 'A':
        votacao_a += 1
    elif voto.upper() == 'B':
        votacao_b += 1
    elif voto.upper() == 'C':
        votacao_c += 1
    else:
        print('Voto nulo')
print(votos)
print(f'CANDIDATO A: {votacao_a}')
print(f'CANDIDATO B: {votacao_b}')
print(f'CANDIDATO C: {votacao_c}')
````

#### 10. Faça um programa que calcule o valor total investido por um colecionador em sua coleção de CDs e o valor médio gasto em cada um deles. O usuário deverá informar a quantidade de CDs e o valor para em cada um.

````python
cds = []
valor_cds = []
quantidade_cds = int(input('Digite a quantidade de cds de sua coleção: '))
for i in range(quantidade_cds):
    nome_cd = input('Digite o nome do cd:')
    cds.append(nome_cd)
    valor_cd = float(input('Digite o valor do cd:'))
    valor_cds.append(valor_cd)
print(cds)
print(f'O valor pago na coleção foi de R${sum(valor_cds)}')
print(f'O valor medio pago por cd foi de R${sum(valor_cds) / quantidade_cds}')
````

#### 14. Uma grande emissora de televisão quer fazer uma enquete entre os seus telespectadores para saber qual o melhor jogador após cada jogo. Sua equipe foi contratada para desenvolver este programa, utilizando a linguagem de programação Python. Para computar cada voto, a telefonista digitará um número, entre 1 e 23, correspondente ao número da camisa do jogador. Um número de jogador igual zero, indica que a votação foi encerrada.         Observe que os votos inválidos e o zero final não devem ser computados como votos. O resultado aparece ordenado pelo número do jogador. O programa deve fazer uso de arrays. Os dados são fictícios e podem mudar a cada execução do programa. Ao final, o programa deve ainda gravar os dados referentes ao resultado da votação em um arquivo texto no disco, obedecendo a mesma disposição apresentada na tela.</pre>
```
Enquete: Quem foi o melhor jogador?
``` 

````python
votos = []
contagens = []
print('Enquete: Quem foi o melhor jogador?')
while True:
    voto = int(input('Número do jogador (0=fim): '))
    if voto < 0 or voto > 23:
        print('Informe um valor entre 1 e 23 ou 0 para sair!')
        continue
    elif voto > 0 and voto < 24:
        votos.append(voto)
        
    else:
        contagem_1 = votos.count(1)
        contagens.append(contagem_1)
        contagem_2 = votos.count(2)
        contagens.append(contagem_2)
        contagem_3 = votos.count(3)
        contagens.append(contagem_3)
        contagem_4 = votos.count(4)
        contagens.append(contagem_4)
        contagem_5 = votos.count(5)
        contagens.append(contagem_5)
        contagem_6 = votos.count(6)
        contagens.append(contagem_6)
        contagem_7 = votos.count(7)
        contagens.append(contagem_7)
        contagem_8 = votos.count(8)
        contagens.append(contagem_8)
        contagem_9 = votos.count(9)
        contagens.append(contagem_9)
        contagem_10 = votos.count(10)
        contagens.append(contagem_10)
        contagem_11 = votos.count(11)
        contagens.append(contagem_11)
        print('Resultado da votação:')
        print('')
        print(f'Foram computados {len(votos)} votos')
        print('')
        print('Jogador   Votos   %')
        for i, contagem in enumerate(contagens):
            if contagem != 0:
                print(f'{i+1}         {contagem}     {(contagem/sum(contagens)):.0%}')
        break
        
````
