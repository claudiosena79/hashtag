# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
salario = np.random.randint(3000, 10500, 48)
meses = np.arange(1,49)

vendas = np.random.randint(1000, 5000, 48)

plt.axis([0, len(meses) + 1, 0, max(salario)+ 500])

plt.xlabel('Meses')
plt.ylabel('R$')

plt.plot(meses, vendas)
plt.show()
```

- Editando o Gráfico de Linha

```python
plt.plot(meses, vendas, color= 'red', linewidth=2.5)
plt.axis([0, len(meses) + 1, 0, max(vendas)+ 500])
plt.show()
```

- Dispersão

```python
plt.scatter(meses, salario, color='green')
plt.xlabel('Meses')
plt.ylabel('R$')
plt.axis([0, len(meses)+1, 0, max(salario)+500])
plt.show()
```

- Barras

```python
plt.bar(meses, salario, color='purple')
plt.xlabel('Meses')
plt.ylabel('R$')
plt.axis([0, len(meses)+1, 0, max(salario)+500])
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(figsize=(12, 15))
plt.subplot(3, 1, 1)
plt.plot(meses, salario, color='red', label='Evolução Salário')
plt.xlabel('Meses')
plt.ylabel('R$')
plt.title('Salário ultimos 48 meses')
plt.axis([0, len(meses)+1, 0, max(salario)+500])
plt.subplot(3, 1, 2)
plt.scatter(meses, salario, color='green')
plt.xlabel('Meses')
plt.ylabel('R$')
plt.axis([0, len(meses)+1, 0, max(salario)+500])
plt.subplot(3, 1, 3)
plt.bar(meses, salario, color='purple')
plt.xlabel('Meses')
plt.ylabel('R$')
plt.axis([0, len(meses)+1, 0, max(salario)+500])
plt.show()
```

- Utilizando OO na construção de Gráficos

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(12, 9))

# Construindo o primeiro Gráfico vendas / meses

axs[0, 0].plot(meses, vendas, color='red', linewidth=1.5)
axs[0, 0].set_title('Vendas x Meses')
axs[0, 0].axis([0, len(meses) + 1, 0, max(vendas) + 500])
axs[0, 1].scatter(meses, vendas, color='green')
axs[0, 1].set_title('Vendas x Meses - Dispersão')
axs[0, 1].axis([0, len(meses) + 1, 0, max(vendas) + 500])
axs[0, 2].bar(meses, vendas, color='purple')
axs[0, 2].set_title('Vendas x Meses - Barras')
axs[1, 0].plot(meses, salario, color='blue', linewidth=1.5)
axs[1, 0].set_title('Salario x Meses')
axs[1, 0].axis([0, len(meses) + 1, 0, max(salario) + 500])
axs[1, 1].scatter(meses, salario, color='orange')
axs[1, 1].axis([0, len(meses) + 1, 0, max(salario) + 500])
axs[1, 1].set_title('Salario x Meses - Dispersão')
axs[1, 2].bar(meses, salario, color='pink')
axs[1, 2].set_title('Salario x Meses - Barras')

plt.tight_layout()
plt.show()
```
