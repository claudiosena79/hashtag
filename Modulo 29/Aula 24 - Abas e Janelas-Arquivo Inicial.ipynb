```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=servico)
```

```python
import os

caminho = os.getcwd()
arquivo = caminho + r"\Pagina Hashtag.html"
driver.get(arquivo)
```

#### Botão que abre outra janela

```python
driver.find_element(By.XPATH,'/html/body/section[2]/div/div[10]/figure/a/img').click()
```

```python
driver.find_element(By.XPATH,'/html/body/section[2]/div/div[4]/figure/a/img').click()
```

```python
driver.find_element(By.NAME,'firstname').send_keys('Claudio Sena')
```

### Outra aba

```python
lista_abas = driver.window_handles # lista com todas as abas abertas pelo selenium

aba_original = driver.window_handles[0]
aba_excel = driver.window_handles[1]
aba_sql = driver.window_handles[2]

for aba in lista_abas:
    driver.switch_to.window(aba)
    print(driver.title)

# Mudando de aba

driver.switch_to.window(aba_sql)
driver.find_element(By.NAME,'firstname').send_keys('Claudio Sena')
driver.find_element(By.ID,'email').send_keys('sql@sql.com')
```
