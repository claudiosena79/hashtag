```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
import os

caminho = os.getcwd()
arquivo = caminho + r"\Pagina Hashtag.html"
navegador.get(arquivo)
```

#### Selecionar pela tag

- Vamos selecionar o t�tulo

```python
from selenium.webdriver.common.by import By

titulo = navegador.find_element(By.TAG_NAME, 'h2').text
print(titulo)
```

### Pegar todos os links das imagens

#### Utilizando Figure

```python
links_imagens =[]
lista_elementos = navegador.find_elements(By.TAG_NAME,'figure')
# print(imagens)
for elemento in lista_elementos:
    try:
        link = elemento.find_element(By.TAG_NAME, 'a').get_attribute('href')
        links_imagens.append(link)
    except:
        continue
print(links_imagens)
```

#### Utilizando Div

```python
links_imagens = []
lista_elementos = navegador.find_elements(By.CLASS_NAME,'wp-block-image')
for elemento in lista_elementos:
    try:
        link = elemento.find_element(By.TAG_NAME, 'a').get_attribute('href')
        links_imagens.append(link)
    except:
        continue
print(links_imagens)
```
